# Data Visualization Practice For Geyser Data set

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load geyser data from seaboarv

```python
geyser = sns.load_dataset('geyser')
```

### first five rows of the geyser data-set

```python
geyser.head()
```

### show the data in a graph

```python
sns.relplot(x='duration', y='waiting', hue=geyser.waiting.tolist(), data=geyser);
```

### group the data by sum and show it in grafh

```python
waiting_sums = geyser.groupby('waiting', as_index=False).sum()
```

```python
sns.lineplot(x='duration', y='waiting', data=waiting_sums);
sns.set(style='darkgrid',)
plt.show()
```

### group the data by average and show it

```python
waiting_mean = geyser.groupby('waiting', as_index=False).mean()
```

```python
sns.lmplot(x='duration', y='waiting',height=5, data=waiting_mean);
# sns.set(style='darkgrid',)
plt.show()
# sns.jointplot("duration", "waiting", data=waiting_mean,
#                   kind="reg", truncate=False,
#                   xlim=(0, 60), ylim=(0, 12),
#                   color="m", height=7)
```
